```python
#광주지역 8월의 최고 기온 히스토그램
import csv
import matplotlib.pyplot as plt

f = open('gwangju.csv', 'r', encoding='cp949')
data = csv.reader(f)
next(data)
aug=[]

for row in data:
    month=row[0].split('.')[1]
    if row[-1]!='':
        if month =='8':
            aug.append(float(row[-1]))
plt.hist(aug,bins=100,color='r')
plt.show()
```

```python
#광주지역 1월과 8월의 최고기온 히스토그램

import csv
import matplotlib.pyplot as plt

f = open('gwangju.csv', 'r', encoding='cp949')
data = csv.reader(f)
next(data)
aug=[]
jan=[]

for row in data:
    month=row[0].split('.')[1]
    if row[-1]!='':
        if month =='8':
            aug.append(float(row[-1]))
        if month =='1':
            jan.append(float(row[-1]))
plt.hist(aug,bins=100,color='r',label='Aug')
plt.hist(jan,bins=100,color='b',label='Jan')
plt.legend()
plt.show()
```

```python
#광주지역 1월과 8월의 상자 그림 히스토그램

import csv
import matplotlib.pyplot as plt

f = open('gwangju.csv', 'r', encoding='cp949')
data = csv.reader(f)
next(data)
aug=[]
jan=[]

for row in data:
    month=row[0].split('.')[1]
    if row[-1]!='':
        if month =='8':
            aug.append(float(row[-1]))
        if month =='1':
            jan.append(float(row[-1]))
plt.boxplot(aug)
plt.boxplot(jan)
plt.show()
```

```python
#광주지역 월별 상자 그림 히스토그램

import csv
import matplotlib.pyplot as plt

f = open('gwangju.csv', 'r', encoding='cp949')
data = csv.reader(f)
next(data)

#월별 데이터를 저장할 리스트 month 생성(12개)
month=[[],[],[],[],[],[],[],[],[],[],[],[]]

for row in data:
    if row[-1]!='':
        #월과 같은 번호의 인덱스에 월별 데이터 저장(예:1월 - month[0])
        month[int(row[0].split('.')[1])-1].append(float(row[-1]))
        
plt.boxplot(month)
plt.show()
```

```python
#광주지역 8월의 일별 상자 그림 히스토그램

import csv
import matplotlib.pyplot as plt

f = open('gwangju.csv', 'r', encoding='cp949')
data = csv.reader(f)
next(data)
day=[]

#일별 데이터를 저장할 리스트 day생성(31개)
for i in range(31):
    day.append([])
    
for row in data:
    if row[-1]!='':
        if row[0].split('.')[1]=='8':
            day[int(row[0].split('.')[2])-1].append(float(row[-1]))

plt.style.use('ggplot')
plt.figure(figsize=(10,5), dpi=300)
plt.boxplot(day,showfliers=False)

plt.show()
```
